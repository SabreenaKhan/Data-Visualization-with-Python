Part2:Data Visualization With Matplotlib

```python
import matplotlib.pyplot as plt
temp=[45,36,32,46,29]
time=[10,12,14,18,23]
plt.figure(figsize=(6,5))
plt.plot(temp,time,color="red")
plt.xlabel("TEMPERATURE")
plt.ylabel("TIME")
plt.title("LINE PLOT")
plt.show()
```

```python
days=["Mon","Tues","Wed","Thurs","Fri","Sat","Sun"]
wt=[20,70,50,60,34,45,87]
plt.figure(figsize=(7,6))
plt.bar(days,wt,color="lavender",edgecolor="black")
plt.xlabel("Week Days")
plt.ylabel("Website Traffic")
plt.title("Bar Graph")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
plt.figure(figsize=(6,5))
data={"marks":np.random.randint(1,101,25)}
df=pd.DataFrame(data)
plt.hist(df["marks"],bins=10,edgecolor="blue")
plt.xlabel("MARKS")
plt.ylabel("FREQUENCY")
plt.title("HISTOGRAM")
plt.show()
```

Section B:Advanced Plots

```python
data={"Weight":[45,55,65,78,86,66,73,21],
      "Height":[153,163,162,187,134,199,157,163]}
df=pd.DataFrame(data)
plt.figure(figsize=(7,6))
plt.scatter(df["Weight"],df["Height"],marker="^",color="blue")
plt.xlabel("Weight")
plt.ylabel("Height")
plt.title("Scatter Plot")
plt.show()
```

```python
data={"company":["Microsoft","Amazon","Oracle","Cisco","Alphabet","IBM","Apple"],
      "share":[20,5,7,15,7,10,36]}
df=pd.DataFrame(data)
plt.pie(df["share"],labels=df["company"],autopct="%1.1f%%")
plt.title("PIE CHART")
plt.show
```

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
img=Image.open(r"C:\Users\Sabreena Khan\Downloads\imagecontour.jpeg").convert("L")
image_data=np.array(img)
plt.figure(figsize=(7,6))
level=np.linspace(1,10,10)
c=plt.contour(image_data,levels=level,cmap="plasma")
plt.imshow(image_data,cmap="gray",alpha=0.5)
plt.colorbar(c,label="Intensity")
plt.tight_layout()
plt.title("IMAGE CONTOUR")
plt.show()
```

Section C:Modern Libraries

```python
wind_direction=[30,60,90,120,150,180,210,240,270,300,330]
wind_speed=[5,7,8,6,4,3,5,7,9,6,4]
theta=np.radians(wind_direction)
plt.figure(figsize=(10,8))
plt.polar(theta,wind_speed,linestyle="-",color="green",marker="*",markersize=10,markerfacecolor="red")
plt.title("Polar Plot")
plt.xlabel("Wind direction")
plt.ylabel("Wind speed")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
months=["Jan","Feb","March","Aplril","May","June"]
Stock_prize=[1200,1000,800,600,900,1500]
plt.figure(figsize=(6,9))
plt.plot(months,Stock_prize,marker="o")
plt.fill_between(months,Stock_prize,color="Red",alpha=0.5)
plt.xlabel("Months")
plt.ylabel("Stock Prizes")
plt.title("Area Plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
population=[1000,2000,3000,4000,5000]
area=[5000,6000,7000,6000,9000]
GDP=[1500,3000,8200,6800,9500]
bs=[]
for i in GDP:
    a=i*1
    bs.append(a)
plt.figure(figsize=(7,6))
plt.scatter(population,area,s=bs,c="pink",alpha=0.8,edgecolor="black")
plt.xlabel("Population")
plt.ylabel("Area")
plt.title("Bubble Chart")
plt.legend()
plt.show()
```

```python
from pywaffle import Waffle
data={"Response":["positive","negative","neutral"],
      "record":[46,54,21]}
df=pd.DataFrame(data)
plt.figure(figsize=(10,9))
fig=plt.figure(FigureClass=Waffle,rows=5,values=df["record"],labels=list(df["Response"]))
plt.title("Waffle Chart")
```
