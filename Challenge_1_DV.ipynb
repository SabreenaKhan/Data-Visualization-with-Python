Here are two coding questions for creating line and bar graphs using pandas, suitable for students who need to upload a CSV file:

**Question 1: Analyzing Weather Data**

**Data:**

The CSV file contains daily weather data for a city, including:

* **Date:** Date of observation (string - format: YYYY-MM-DD)
* **Temperature (High):** Maximum temperature for the day (integer)

**Tasks:**

1. **Import libraries:** Begin by importing `pandas` for data manipulation and `matplotlib.pyplot` for creating visualizations.
2. **Read CSV:** Use `pandas.read_csv` to read the data from the uploaded CSV file.
3. **Line plot:** Create a line plot to visualize the **daily high temperatures** over time.
    * X-axis: Date (converted to desired format)
    * Y-axis: Temperature (High)

**Data Example (weather.csv):**

```csv
Date,Temperature (High)
2023-12-01,15
2023-12-02,18
2023-12-03,12
2023-12-04,10
2023-12-05,17
```

**Question 2: Movie Genre Popularity**

**Data:**

The CSV file contains information about movies, including:

* **Title:** Movie title (string)
* **Genre:** Movie genre (string)

**Tasks:**

1. **Import libraries:** Similar to Question 1.
2. **Read CSV:** Use `pandas.read_csv` to read the data from the uploaded CSV file.
3. **Bar graph:** Create a bar graph to show the **distribution of movies by genre**.
    * X-axis: Genre
    * Y-axis: Number of movies in each genre (count)

**Data Example (movies.csv):**

```csv
Title,Genre
The Shawshank Redemption,Drama
The Godfather,Crime
The Dark Knight,Action,Thriller
The Godfather: Part II,Crime
Pulp Fiction,Crime,Comedy
The Lord of the Rings: The Return of the King,Fantasy,Adventure
Schindler's List,Historical Drama
12 Angry Men,Drama
The Lord of the Rings: The Fellowship of the Ring,Fantasy,Adventure
Fight Club,Drama
```

**Code Template (applicable to both questions):**

````python
import matplotlib.pyplot as plt
import pandas as pd
data={"Date":["01-12-2023","02-12-2023","03-12-2023","04-12-2023","05-12-2023"],
      "Temperature(High)":[15,18,12,10,17]}
df=pd.DataFrame(data)
df
````

````python
plt.plot(df["Date"],df["Temperature(High)"])
plt.xlabel("Date")
plt.ylabel("Maximum Temperature")
plt.title("Weather Report")
````
