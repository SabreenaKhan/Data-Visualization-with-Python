take the input from the user and check if the value is a prime no or not.

```python
num=int(input("Enter a number:"))
if(num==1):
    print("Not a prime number")
c = False
for i in range(2,num):
    if num%i==0:
        c = True
        break
    else:
        c = False

if c == False:
    print("Prime")
else:
    print("Not Prime")

    

```

box plot

```python
import pandas as pd
data={"col":[78,82,85,88,90,92,93,95,97,100,65,72,150,40]}
df=pd.DataFrame(data)
```

```python

per25=df["col"].quantile(0.25)
per75=df["col"].quantile(0.75)
iqr=(per75-per25)
ul=per75+(iqr*1.5)
ll=per25-(iqr*1.5)
df=df[df["col"]<=ul]
df=df[df["col"]>=ll]
plt.boxplot(df)
df

```

3d scatter plot

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
x=[1,2,3,4,5]
y=[4,6,5,8,2]
z=[7,3,1,5,2]
fig=plt.figure(figsize=(10,8))
ax=fig.add_subplot(111,projection="3d")
ax.scatter(x,y,z,marker="*")
ax.set_xlabel("X Data")
ax.set_ylabel("Y Data")
ax.set_zlabel("Z Data")
plt.title("simple 3d scatter plot")
plt.show()


```
