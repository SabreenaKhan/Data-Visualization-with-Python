

**Question 1: Country Life Expectancy vs. GDP**

**Data:**

The CSV file contains information about countries:

* **Country:** Name of the country (string) - **Do not use for plotting**
* **Life Expectancy:** Average life expectancy at birth (float)
* **GDP per Capita (USD):** Gross Domestic Product per capita (float)

**Task:** Create a scatter plot to explore the relationship between a country's **GDP per capita** (X-axis) and its **life expectancy** (Y-axis).

**Data Example (world_data.csv):**

```csv
Country,Life Expectancy,GDP per Capita (USD)
Japan,84.3,39522
Switzerland,84.2,87169
Singapore,83.7,65870
Australia,82.8,51408
Iceland,83.3,67004
```

**Colab Code:**

````python
import matplotlib.pyplot as plt
import pandas as pd
data={"Country":["japan","switzerland","singapore","australia","iceland"],
"Life Expectancy":[84.3,84.2,83.7,82.8,83.3],
"GDP Capita(USD)":[39522,87169,65870,51408,67004]}
df=pd.DataFrame(data)
plt.scatter(df["GDP Capita(USD)"],df["Life Expectancy"],marker="*",color="pink",edgecolor="black")
plt.title("Relationship between a country's GDP and its Life Expectancy")
plt.xlabel("GDP per capita")
plt.ylabel("Life Expectancy")
plt.show()
````

**Question 2: Flight Distance vs. Travel Time**

**Data:**

Imagine you have data on flights between different cities:

* **Origin City:** City of departure (string) - **Do not use for plotting**
* **Destination City:** City of arrival (string) - **Do not use for plotting**
* **Distance (km):** Distance between the two cities (float)
* **Travel Time (hours):** Flight travel time (float)

**Task:** Create a scatter plot to visualize the relationship between the **flight distance** (X-axis) and the **travel time** (Y-axis).

**Data Example:** (This data can be hypothetical and doesn't require a CSV file)

| Origin City | Destination City | Distance (km) | Travel Time (hours) |
|---|---|---|---|
| New York | London | 5575 | 7.0 |
| Los Angeles | Tokyo | 8974 | 11.5 |
| Paris | Beijing | 8563 | 10.0 |
| Dubai | Sydney | 12844 | 16.0 |
| Frankfurt | Singapore | 10621 | 13.5 |

**Colab Code:**

````python
import matplotlib.pyplot as plt
import pandas as pd
data={"Origin City":["New York","Los Angeles","Paris","Dubai","Frankfurt"],
"Destination City":["London","Tokyo","Beijing","Sydney","singapore"],
"Distance(km)":[5575,8974,8563,12844,10621],
"Travel Time(hours)":[7.0,11.5,10.0,16.0,13.5]}
df=pd.DataFrame(data)
plt.scatter(df["Distance(km)"],df["Travel Time(hours)"],marker="*",color="red",edgecolor="black")
plt.title("Relationship between Flight distance and travel time")
plt.xlabel("Flight Distance")
plt.ylabel("Travel Time")
plt.show()
````

**Question 3: Employee Salary vs. Years of Experience**

**Data:**

A CSV file contains information about company employees:

* **Employee Name:** Name of the employee (string) - **Do not use for plotting**
* **Years of Experience:** Total years of experience (integer)
* **Salary (USD):** Annual salary (float)

**Task:** Create a scatter plot to explore the relationship between the **years of experience** (X-axis) and the **employee salary** (Y-axis).

**Data Example (employee_data.csv):**

```csv
Employee Name,Years of Experience,Salary (USD)
John Smith,5,75000
Jane Doe,8,92000
Michael Lee,3,58000
Alice Johnson,10,1100

````python
import matplotlib.pyplot as plt
import pandas as pd
data={"Employee Name":["John Smith","Jane Doe","Michael Lee","Alice Johnson"],
"Years of Experience":[5,8,3,10],"Salary(USD)":[75000,92000,58000,110000]}
df=pd.DataFrame(data)
plt.scatter(df["Years of Experience"],df["Salary(USD)"],marker="*",color="pink",edgecolor="black")
plt.title("Relationship between years of experience and the employee salary")
plt.xlabel("Years of Experience")
plt.ylabel("Employee Salary")
plt.show()
````
