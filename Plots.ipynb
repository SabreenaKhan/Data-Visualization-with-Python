IMAGE CONTOUR

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
img=Image.open(r"C:\Users\Sabreena Khan\Downloads\pandaphoto.jpeg").convert("L")
image_data=np.array(img)
plt.figure(figsize=(10,8))
level=np.linspace(1,10,10)
c=plt.contour(image_data,levels=level,cmap="plasma")
plt.imshow(image_data,cmap="gray",alpha=0.5)
plt.colorbar(c,label="Intensity")
plt.tight_layout()
plt.show()
```

POLAR PLOT

```python
wd=[30,60,90,120,150,180,210,240,270,300,330]
ws=[5,7,8,6,4,3,5,7,9,6,4]
theta=np.radians(wd)
plt.figure(figsize=(10,8))
plt.polar(theta,ws,linestyle="-",color="green",marker="*",markersize=10,markerfacecolor="red")
plt.title("Polar Plot")
plt.xlabel("Wind direction")
plt.ylabel("Wind speed")
plt.show()
```

BOX PLOT

```python
import pandas as pd
data={"col":[78,82,85,88,90,92,93,95,97,100,65,72,150,40]}
df=pd.DataFrame(data)
per25=df["col"].quantile(0.25)
per75=df["col"].quantile(0.75)
iqr=(per75-per25)
ul=per75+(iqr*1.5)
ll=per25-(iqr*1.5)
df=df[df["col"]<=ul]
df=df[df["col"]>=ll]
plt.boxplot(df)
df
```

BUBBLE PLOT

```python
import matplotlib.pyplot as plt
x=[10,20,30,40,50]
y=[5000,6000,7000,6000,9000]
z=[15,30,82,68,95]
bs=[]
for i in z:
    a=i*50
    bs.append(a)
plt.figure(figsize=(8,6))
plt.scatter(x,y,s=bs,c="yellow",alpha=0.8,edgecolor="black")
plt.xlabel("population")
plt.ylabel("income")
plt.title("Bubble plot")
plt.legend()
plt.show()
```

WAFFLE CHART

```python
import matplotlib.pyplot as plt
import pandas as pd
from pywaffle import Waffle
data={"Phone":["Mi","Apple","Samsung","Nokia","Realme"],
      "Stocks":[44,12,8,5,3]}
df=pd.DataFrame(data)
fig=plt.figure(figsize=(8,9),FigureClass=Waffle,rows=5,values=df.Stocks,label=list(df.Phone))
plt.title("Waffle Chart")
plt.legend()
plt.show()
```

AREA PLOT

```python
import matplotlib.pyplot as plt
import pandas as pd
months=["Jan","Feb","March","Aplril","May","June"]
Sales=[1200,1000,800,600,900,1500]
plt.figure(figsize=(6,9))
plt.plot(months,Sales,marker="o")
plt.fill_between(months,Sales,color="Red",alpha=0.5)
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("sales following months")
plt.show()
```
